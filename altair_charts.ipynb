```python
import altair as alt
import pandas as pd

# 1. Create Data
data = pd.DataFrame({
    'cat': ['A', 'B', 'C', 'D'],
    'val': [30, 80, 45, 60]
})

# 2. Build Charts
bars = alt.Chart(data).mark_bar().encode(
    x='cat:N',
    y='val:Q',
    tooltip=['cat', 'val']
)

lines = alt.Chart(data).mark_line(color='red').encode(
    x='cat:N',
    y='val:Q'
)

# 3. Layer and Display
(bars + lines).properties(width=400, height=300).interactive()
```
